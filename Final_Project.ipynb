```python
from datetime import datetime, timedelta
def convert_str_to_minutes(time_str):
  if(isinstance(time_str, str)):
    hours, minutes, seconds = map(int, time_str.split(':'))
    total_minutes = hours * 60 + minutes + seconds / 60
    return int(total_minutes)
  else:
    return time_str
```

```python
# Define a function to convert Chinese time to 24-hour format
def convert_chinese_time(time_str):
    # Determine if it's AM or PM
    is_pm = "下午 " in time_str

    # Remove "上午" or "下午" from the string
    time_str = time_str.replace("上午", "AM").replace("下午", "PM")

    # Convert to datetime object
    time_format = "%p %I:%M:%S" #if not is_pm else "%p %I:%M:%S"
    time_object = datetime.strptime(time_str, time_format)

    return time_object
```

```python
df['Alarm time'] = df['Alarm time'].apply(convert_chinese_time)
df['Time I sit on my car'] = df['Time I sit on my car'].apply(convert_chinese_time)
df['Class starting time'] = df['Class starting time'].apply(convert_chinese_time)
df['Arrival time'] = df['Arrival time'].apply(convert_chinese_time)
df['Arrival time'].head(5)
```

```python
col2use = ['Alarm time_hour','Alarm time_minute','Days of the week','Weather forcast at last night','Temperature forcast (celsius)','Probability of precipitation forcast','Class starting time_hour','Class starting time_minute']
col2get = ['Arrival time_hour','Arrival time_minute']
df_test=df[col2use + col2get]
print(df_test.dtypes)
df_temp01=df_test.copy()
df_temp02=df_test.copy()
df_temp03=df_test.copy()
df_temp11=df_test.copy()
df_temp12=df_test.copy()
df_temp13=df_test.copy()
df_temp01['Alarm time_minute'] = pd.to_numeric(df_temp01['Alarm time_minute'], errors='coerce') + 1
df_temp01['Arrival time_minute']=pd.to_numeric(df_temp01['Arrival time_minute'], errors='coerce') + 1
df_temp02['Alarm time_minute'] = pd.to_numeric(df_temp02['Alarm time_minute'], errors='coerce') + 2
df_temp02['Arrival time_minute']=pd.to_numeric(df_temp02['Arrival time_minute'], errors='coerce') + 2
df_temp03['Alarm time_minute']=pd.to_numeric(df_temp03['Alarm time_minute'], errors='coerce') + 3
df_temp03['Arrival time_minute']=pd.to_numeric(df_temp03['Arrival time_minute'], errors='coerce') + 3
df_temp11['Alarm time_minute']=pd.to_numeric(df_temp11['Alarm time_minute'], errors='coerce') -1
df_temp11['Arrival time_minute']=pd.to_numeric(df_temp11['Arrival time_minute'], errors='coerce') - 1
df_temp12['Alarm time_minute']=pd.to_numeric(df_temp12['Alarm time_minute'], errors='coerce') - 2
df_temp12['Arrival time_minute']=pd.to_numeric(df_temp12['Arrival time_minute'], errors='coerce') - 2
df_temp13['Alarm time_minute']=pd.to_numeric(df_temp13['Alarm time_minute'], errors='coerce') - 3
df_temp13['Arrival time_minute']=pd.to_numeric(df_temp13['Arrival time_minute'], errors='coerce') - 3
df_train=pd.concat([df_temp01,df_temp02,df_temp03,df_temp11,df_temp12,df_temp13])
df_train.head(5)
```

**Build ML model**

```python
X_train = df_train[col2use].values
X_test = df_test[col2use].values
y_train = df_train[col2get].values
y_test = df_test[col2get].values
print('Training shapes:',X_train.shape, y_train.shape)
print('Validation shapes:',X_test.shape, y_test.shape)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.metrics import classification_report, confusion_matrix
np.seterr(divide='ignore', invalid='ignore', over='ignore')
def model_prediction(model):
    model.fit(X_train,y_train)
    X_train_pred = model.predict(X_train)
    X_test_pred = model.predict(X_test)

    n = X_test.shape[0]
    p = X_test.shape[1]
    r2_adj=1-(1-r2_score(y_test, X_test_pred))*(n-1)/(n-p-1)


    # The mean squared error
    print("Root mean squared error: %.2f"
          % sqrt(mean_squared_error(y_test, X_test_pred)))
    # The absolute squared error
    print("Mean absolute error: %.2f"
          % mean_absolute_error(y_test, X_test_pred))
    # Explained variance score: 1 is perfect prediction
    print('R-squared: %.2f' % r2_score(y_test, X_test_pred))

    # Explained adjusted r^2:
    print('Adjusted R-squared: %.2f' % r2_adj)
```

```python
model_prediction(RandomForestRegressor(n_estimators = 1))
```

```python
md=RandomForestRegressor(n_estimators = 1).fit(X_train,y_train)

def alarm_time_optimizer():
  # set allowance
  delta_time=55
  allowance=3
  itertime=0

  #user enter condition
  days_of_the_week = input("Enter Days of the Week: ")
  weather = input("Enter Weather: ")
  temperature = input("Enter Temperature: ")
  probability_of_precipitation = input("Enter Probability of Precipitation: ")
  class_start_time_hour = int(input("Enter Class Start Time Hour: "))
  class_start_time_minute = int(input("Enter Class Start Time Minute: "))

  #generate initial alarm time
  alarm_hour = class_start_time_hour-2
  alarm_minute = 1

  # Create a dictionary with user input
  user_data = {
      'Alarm_hour': [alarm_hour],
      'Alarm_minute': [alarm_minute],
      'Days_of_the_week': [days_of_the_week],
      'Weather': [weather],
      'Temperature': [temperature],
      'Probability_of_precipitation': [probability_of_precipitation],
      'Class_start_time_hour': [class_start_time_hour],
      'Class_start_time_minute': [class_start_time_minute]
  }
  # Create a DataFrame from the dictionary
  df = pd.DataFrame(user_data)
  while(delta_time>allowance and itertime<100):
    itertime=itertime+1
    prediction=md.predict(df)
    delta_time=(df['Class_start_time_hour'][0]-float(prediction[0,0]))*60+(df['Class_start_time_minute'][0]-float(prediction[0,1]))
    df['Alarm_minute']=df['Alarm_minute']+1
    if df['Alarm_minute'][0]>=60:
      df['Alarm_minute']=df['Alarm_minute']-60
      df['Alarm_hour'][0]=df['Alarm_hour']+1
  prediction=md.predict(df)
  print('The ideal alarm time is:',df['Alarm_hour'][0],':',df['Alarm_minute'][0])
  print('You will arrive at:',int(prediction[0,0]),':',int(prediction[0,1]))
```
